```python
import matplotlib.pyplot as plt

data['MP'] = pd.to_numeric(data['MP'], errors='coerce')
data['PTS'] = pd.to_numeric(data['PTS'], errors='coerce')

plt.figure(figsize=(12, 8))
plt.hexbin(data['Age'], data['MP'], gridsize=30, cmap='viridis')
plt.colorbar(label='Frequency')
plt.xlabel('Matches Play')
plt.ylabel('Points (PTS)')
plt.show()
```

```python
import seaborn as sns

data['PTS'] = pd.to_numeric(data['PTS'], errors='coerce')
data['AST'] = pd.to_numeric(data['AST'], errors='coerce')

plt.figure(figsize=(10, 8))
sns.kdeplot(x=data['PTS'], y=data['AST'], cmap='plasma', fill=True, thresh=0.05)
plt.title('Density Plot of Points (PTS) vs Assists (AST)')
plt.xlabel('Points (PTS)')
plt.ylabel('Assists (AST)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data['FT%'] = pd.to_numeric(data['FT%'], errors='coerce')
data['3P%'] = pd.to_numeric(data['3P%'], errors='coerce')

plt.figure(figsize=(12, 8))
sns.scatterplot(x='FT%', y='3P%', data=data, hue='Age', palette='coolwarm', size='PTS', sizes=(20, 200), alpha=0.7)
plt.title('Free Throw % vs 3-Point % by Age and Points')
plt.xlabel('Free Throw % (FT%)')
plt.ylabel('3-Point % (3P%)')
plt.legend(title='Age')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
data['MP'] = pd.to_numeric(data['MP'], errors='coerce')

plt.figure(figsize=(12, 8))
sns.boxplot(x='Age', y='MP', data=data, palette='Spectral')
plt.title('Minutes Played (MP) Distribution by Age')
plt.xlabel('Age')
plt.ylabel('Minutes Played (MP)')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

data['G'] = pd.to_numeric(data['G'], errors='coerce')
data['PTS'] = pd.to_numeric(data['PTS'], errors='coerce')

plt.figure(figsize=(12, 8))
sns.regplot(x='G', y='PTS', data=data, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Games Played vs Points Scored')
plt.xlabel('Games Played (G)')
plt.ylabel('Points Scored (PTS)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data['AST'] = pd.to_numeric(data['AST'], errors='coerce')
data['TOV'] = pd.to_numeric(data['TOV'], errors='coerce')

plt.figure(figsize=(12, 8))
sns.jointplot(x='AST', y='TOV', data=data, kind='hex', gridsize=30, cmap='Blues')
plt.suptitle('Hexbin Plot of Assists (AST) vs Turnovers (TOV)', y=1.02)
plt.xlabel('Assists (AST)')
plt.ylabel('Turnovers (TOV)')
plt.show()
```
